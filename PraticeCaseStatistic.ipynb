```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Fellowship-main/Seasons_Stats.csv')
```

# Preprocessing

```python
df = df[df['Year'] == 2017]
df.info()
```

```python
df = df.drop(['Unnamed: 0', 'blanl', 'blank2'], axis=1)
df = df.drop_duplicates(subset='Player', keep="last")
```

```python
dfna = df[df.isna().any(axis=1)]
dfna
```

```python
df = df.fillna(0)
df.isnull().values.any()
```

# Answer 1


```python
age = df.groupby(['Tm']).agg({'Age':['min','max']})
```

```python
age_np = age.to_numpy()
age_indx = age.index.to_numpy()
df_np = df.to_numpy()
```

```python
def name_age(d):
  a = []
  for i in range(0,age_indx.shape[0]):
      for j in range (0,df.shape[0]):
        if df_np[j,4] == age_indx[i] and df_np[j,3] == age_np[i,d]:
          a.append([age_indx[i],df_np[j,1],age_np[i,d]])
  return a
```

```python
name_min = name_age(0)
name_max = name_age(1)
```

```python
age_indx = age_indx.reshape((age_indx.shape[0],-1))
```

```python
age_max = pd.DataFrame(name_max, columns = ['Team', 'Name', 'Age'])
age_max
```

```python
age_min = pd.DataFrame(name_min, columns = ['Team', 'Name', 'Age'])
age_min
```

# Answer 2

```python
mp = df.groupby(['Pos']).agg({'MP':'max'})
```

```python
mp_np = mp.to_numpy()
mp_indx = mp.index.to_numpy()
```

```python
def name_mp(d):
  a = []
  for i in range(0,mp_indx.shape[0]):
      for j in range (0,df.shape[0]):
        if df_np[j,2] == mp_indx[i] and df_np[j,7] == mp_np[i,d]:
          a.append([mp_indx[i],df_np[j,1],mp_np[i,d]])
  return a
```

```python
mp_max = name_mp(0)
mp_max_df = pd.DataFrame(mp_max, columns = ['Pos', 'Name', 'MP'])
```
